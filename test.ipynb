```python

def scatter_plot(x: np.ndarray, y: np.ndarray, 
                 output_location: Path, 
                 plot_title: str,
                 xlabel: str, 
                 ylabel: str) -> None:
    
    plt.figure(figsize=(10, 8))
    plt.scatter(x, y, alpha=0.5)

    plt.plot([-5, 5], [-5, 5], color='grey', linestyle='--', label='y=x line')
    # add a pearson correlation coefficient to the plot

    # calculate Pearson correlation coefficient with nan values removed
    mask = ~np.isnan(x) & ~np.isnan(y)
    x = x[mask]
    y = y[mask]

    correlation = np.corrcoef(x, y)[0, 1]
    plt.text(0.05, 0.95, f'Pearson r = {correlation:.2f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(plot_title)
    plt.show()



x_vals = df_old.loc['DDR1', :]
y_vals = transcriptome.loc['DDR1', :]
x_vals.sort_index(inplace=True)
y_vals.sort_index(inplace=True)

scatter_plot(x_vals.values, y_vals.values, f'scatter_plot_zoom.png', plot_title=f'', xlabel=f'protein zscore', ylabel=f'transcriptome zscore')
```
